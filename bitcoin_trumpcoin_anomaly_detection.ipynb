# Gloss√°rio T√©cnico

## A

**Anomalia**
Evento raro, inesperado ou extremo que foge do comportamento normal dos dados. Em mercados financeiros, pode indicar manipula√ß√£o, crashes ou movimentos at√≠picos.

**Aprendizado N√£o Supervisionado**
Tipo de Machine Learning em que o modelo aprende padr√µes sem r√≥tulos pr√©vios. Muito usado quando n√£o existe *ground truth* confi√°vel.

## B

**Bandas de Bollinger**
Indicador estat√≠stico formado por uma m√©dia m√≥vel e faixas superior e inferior baseadas em desvio padr√£o. Pre√ßos fora das bandas indicam poss√≠veis anomalias.

## C

**Clipping**
T√©cnica de pr√©-processamento que limita valores extremos a determinados quantis, reduzindo a influ√™ncia de outliers no modelo.

**Close (C)**
Pre√ßo de fechamento de um ativo em determinado per√≠odo.

**Contamination / Nu**
Par√¢metro que define a propor√ß√£o esperada de anomalias no conjunto de dados (ex.: 1‚Äì1,5%).

## D

**Desbalanceamento de Classes**
Situa√ß√£o em que a classe normal √© muito maior que a classe an√¥mala, comum em detec√ß√£o de fraudes e anomalias.

**Detec√ß√£o de Anomalias**
Conjunto de t√©cnicas voltadas √† identifica√ß√£o de padr√µes raros ou extremos em dados.

## F

**Feature (Atributo)**
Vari√°vel utilizada pelo modelo para aprender padr√µes (ex.: pre√ßo, volume, RSI).

**Feature Engineering**
Processo de cria√ß√£o e transforma√ß√£o de vari√°veis para melhorar o desempenho do modelo.

**Flash Crash**
Queda abrupta e intensa de pre√ßo em curto intervalo de tempo.

**F1-score**
M√©dia harm√¥nica entre precision e recall, usada para avaliar modelos em dados desbalanceados.

## G

**Ground Truth**
Refer√™ncia considerada como ‚Äúverdade‚Äù. Em projetos n√£o supervisionados, costuma ser apenas uma aproxima√ß√£o estat√≠stica.

## H

**High (H)**
Pre√ßo m√°ximo de um ativo em determinado per√≠odo.

## I

**IA (Intelig√™ncia Artificial)**
√Årea da computa√ß√£o que desenvolve sistemas capazes de aprender padr√µes e tomar decis√µes de forma aut√¥noma.

**Isolation Forest**
Algoritmo de detec√ß√£o de anomalias baseado no isolamento r√°pido de pontos raros por meio de √°rvores aleat√≥rias.

## L

R√≥tulos artificiais criados com regras estat√≠sticas para avalia√ß√£o do modelo, n√£o para treinamento.

**Low (L)**
Pre√ßo m√≠nimo de um ativo em determinado per√≠odo.

Algoritmo que detecta anomalias comparando a densidade local de um ponto com a de seus vizinhos.

## M

**MACD (Moving Average Convergence Divergence)**
Indicador de tend√™ncia baseado na diferen√ßa entre m√©dias m√≥veis exponenciais.

**Machine Learning**
Sub√°rea da IA focada em modelos que aprendem padr√µes estat√≠sticos automaticamente a partir de dados.

## N

**NaN (Not a Number)**
Valor ausente ou indefinido em um conjunto de dados.

**Normaliza√ß√£o**
Processo de padroniza√ß√£o das vari√°veis para que fiquem na mesma escala.

## O

Conjunto de pre√ßos: Open, High, Low e Close. Base da an√°lise de s√©ries temporais financeiras.

**One-Class SVM**

**Open (O)**
Pre√ßo de abertura de um ativo em determinado per√≠odo.

## Instalar Depend√™ncias

```python
# Instalar depend√™ncias
!pip install yfinance -q

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import precision_score, recall_score, f1_score, classification_report
import yfinance as yf
import warnings
warnings.filterwarnings('ignore')

plt.style.use('default')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)
```

## Fun√ß√µes Gerais

```python
# FUN√á√ïES T√âCNICAS PROFISSIONAIS
def calculate_rsi(series, period=14):
    """RSI: detecta sobrecompra (>70) e sobrevenda (<30)"""
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

def calculate_macd(series, fast=12, slow=26):
    """MACD: momentum e revers√µes de tend√™ncia"""
    ema_fast = series.ewm(span=fast).mean()
    ema_slow = series.ewm(span=slow).mean()
    return ema_fast - ema_slow

def calculate_bollinger_upper(series, window=20, num_std=2):
    """Bollinger Upper: detecta rompimentos explosivos"""
    rolling_mean = series.rolling(window).mean()
    rolling_std = series.rolling(window).std()
    return rolling_mean + (rolling_std * num_std)

def advanced_features(df):
    """Feature engineering avan√ßado para anomalias"""
    df = df.copy()

    # Indicadores b√°sicos
    df['RSI'] = calculate_rsi(df['Close'])
    df['MACD'] = calculate_macd(df['Close'])
    df['BB_upper'] = calculate_bollinger_upper(df['Close'])

    # Volume e volatilidade
    df['Volume_Change'] = df['Volume'].pct_change()
    df['Volatility'] = df['Close'].rolling(24).std()
    df['Price_Change'] = df['Close'].pct_change()

    # Features avan√ßadas de anomalia
    df['RSI_Extreme'] = (df['RSI'] > 80) | (df['RSI'] < 20)  # Sobrecompra/sobrevenda extrema
    df['Volume_Spike'] = df['Volume_Change'] > df['Volume_Change'].quantile(0.95)
    df['Price_Gap'] = abs(df['Price_Change']) > df['Price_Change'].quantile(0.95)

    # Features temporais
    df['Hour'] = df.index.hour
    df['DayOfWeek'] = df.index.dayofweek

    return df
```

## Detec√ß√£o de Anomalias Bitcoin

```python
print("=== DETEC√á√ÉO DE ANOMALIAS BITCOIN - VERS√ÉO PROFISSIONAL ===\n")
btc_raw = yf.download('BTC-USD', period='1y', interval='1h')
btc = advanced_features(btc_raw)
btc.dropna(inplace=True)

print(f"Dataset: {btc.shape}")
print(f"Per√≠odo: {btc.index.min():%Y-%m-%d} at√© {btc.index.max():%Y-%m-%d}")
print("\nFeatures criadas:")
print([col for col in btc.columns if col not in btc_raw.columns])
```

## Pr√©-processamento + Labels

```python
# Features num√©ricas para ML (excluindo booleanas temporariamente)
features = ['Close', 'Volume', 'RSI', 'MACD', 'BB_upper', 'Volume_Change',
           'Volatility', 'Hour', 'DayOfWeek']
X_btc = btc[features].copy()

print("=== TRATAMENTO DE DADOS ===")
print("1. Removendo infinitos...")
X_btc.replace([np.inf, -np.inf], np.nan, inplace=True)

print("2. Preenchendo NaN...")
X_btc.fillna(0, inplace=True)

print("3. Clipping extremos...")
for col in X_btc.columns:
    lower = X_btc[col].quantile(0.01)
    upper = X_btc[col].quantile(0.99)
    X_btc[col] = X_btc[col].clip(lower, upper)

# Labels H√çBRIDOS (mais realistas)
btc['Anomaly_Label'] = (
    (abs(btc['Price_Change']) > abs(btc['Price_Change']).quantile(0.99)) |  # Flash crash/pump
    (btc['RSI'] > 85) | (btc['RSI'] < 15) |                                # RSI extremo
    (btc['Volume_Change'] > btc['Volume_Change'].quantile(0.98))            # Volume spike
).astype(int)

# Normaliza√ß√£o robusta (insens√≠vel a outliers)
scaler = RobustScaler()
X_btc_scaled = scaler.fit_transform(X_btc)

print(f"\nFeatures finais: {X_btc_scaled.shape}")
print(f"Anomalias reais: {btc['Anomaly_Label'].sum()} ({btc['Anomaly_Label'].mean():.1%})")
print("‚úÖ Pr√©-processamento conclu√≠do!")
```

## Treinamento dos Algoritmos (4 Algoritmos)

```python
print("=== TREINAMENTO DOS 4 MODELOS ===")

# 4 algoritmos (3 obrigat√≥rios + bonus)
models = {
    'Isolation Forest': IsolationForest(contamination=0.015, random_state=42, n_jobs=-1),
    'Local Outlier Factor': LocalOutlierFactor(n_neighbors=25, contamination=0.015),
    'One-Class SVM': OneClassSVM(nu=0.015, kernel='rbf', gamma='scale'),
    'Isolation Forest (Deep)': IsolationForest(contamination=0.015, max_samples=0.8, random_state=42)
}

predictions = {}
for name, model in models.items():
    print(f"Treinando {name}...", end=" ")
    preds = model.fit_predict(X_btc_scaled)
    anomalies = np.where(preds == -1)[0]
    predictions[name] = anomalies
    print(f"{len(anomalies)} anomalias")

print("\n‚úÖ Todos os modelos treinados!")
```

### M√©tricas

```python
from google.colab import files

print("=== AVALIA√á√ÉO DETALHADA ===")
y_true = btc['Anomaly_Label'].values

results = []
for name, anomalies in predictions.items():
    y_pred = np.zeros(len(btc))
    if len(anomalies) > 0:
        y_pred[anomalies] = 1

    prec = precision_score(y_true, y_pred, zero_division=0)
    rec = recall_score(y_true, y_pred, zero_division=0)
    f1 = f1_score(y_true, y_pred, zero_division=0)

    results.append({
        'Algoritmo': name,
        'Anomalias': len(anomalies),
        'Precision': prec,
        'Recall': rec,
        'F1-Score': f1
    })

    print(f"{name}: Precision={prec:.3f}, Recall={rec:.3f}, F1={f1:.3f}")

# Tabela final
results_df = pd.DataFrame(results).round(3)
print("\nTABELA FINAL:")
print(results_df)
```
