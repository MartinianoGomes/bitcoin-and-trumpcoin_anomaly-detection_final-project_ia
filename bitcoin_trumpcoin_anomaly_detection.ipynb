# Algoritmos de Detec√ß√£o de Anomalias Utilizados

## 1. Isolation Forest

![Isolation Forest](https://scikit-learn.org/stable/_images/sphx_glr_plot_isolation_forest_001.png)

### Como Funciona

O **Isolation Forest** √© baseado em uma ideia simples e elegante: **anomalias s√£o mais f√°ceis de isolar** do que pontos normais.

O algoritmo constr√≥i m√∫ltiplas **√°rvores de decis√£o aleat√≥rias** (Random Trees) que particionam o espa√ßo de dados recursivamente:

1. **Sele√ß√£o aleat√≥ria** de uma feature (atributo)
2. **Sele√ß√£o aleat√≥ria** de um ponto de corte entre o m√≠nimo e m√°ximo dessa feature
3. **Repeti√ß√£o** at√© que cada ponto esteja isolado ou atinja a profundidade m√°xima

### Intui√ß√£o

- **Anomalias** (pontos raros/extremos) s√£o isoladas rapidamente ‚Üí **caminho curto** na √°rvore
- **Pontos normais** (densos) precisam de mais parti√ß√µes ‚Üí **caminho longo** na √°rvore

O **Anomaly Score** √© calculado pela profundidade m√©dia de isolamento em todas as √°rvores.

### Par√¢metros Principais

| Par√¢metro | Valor Usado | Descri√ß√£o |
|-----------|-------------|-----------|
| `contamination` | 0.015 | Propor√ß√£o esperada de anomalias (~1.5%) |
| `n_estimators` | 100 (default) | N√∫mero de √°rvores no ensemble |
| `random_state` | 42 | Semente para reprodutibilidade |

### Vantagens
‚úÖ Muito r√°pido e escal√°vel (complexidade O(n log n))  
‚úÖ Funciona bem em alta dimensionalidade  
‚úÖ N√£o assume distribui√ß√£o espec√≠fica dos dados

## 2. Local Outlier Factor (LOF)

![Local Outlier Factor](https://scikit-learn.org/stable/_images/sphx_glr_plot_lof_outlier_detection_001.png)

### Como Funciona

O **LOF (Local Outlier Factor)** detecta anomalias comparando a **densidade local** de um ponto com a densidade de seus vizinhos mais pr√≥ximos.

O algoritmo funciona em 4 etapas:

1. **k-dist√¢ncia**: Para cada ponto, calcula a dist√¢ncia at√© o k-√©simo vizinho mais pr√≥ximo
2. **Alcan√ßabilidade**: Define a "dist√¢ncia de alcan√ßabilidade" entre pares de pontos
3. **Densidade Local (LRD)**: Calcula a densidade local de cada ponto baseada na alcan√ßabilidade m√©dia
4. **LOF Score**: Raz√£o entre a densidade dos vizinhos e a densidade do pr√≥prio ponto

### Intui√ß√£o

- **LOF ‚âà 1**: Ponto tem densidade similar aos vizinhos ‚Üí **Normal**
- **LOF >> 1**: Ponto tem densidade muito menor que vizinhos ‚Üí **Anomalia**
- **LOF << 1**: Ponto mais denso que vizinhos (raro)

### Par√¢metros Principais

| Par√¢metro | Valor Usado | Descri√ß√£o |
|-----------|-------------|-----------|
| `n_neighbors` | 25 | N√∫mero de vizinhos para calcular densidade local |
| `contamination` | 0.015 | Propor√ß√£o esperada de anomalias (~1.5%) |

### Vantagens
‚úÖ Detecta anomalias **locais** (em regi√µes de diferentes densidades)  
‚úÖ N√£o assume formato espec√≠fico dos clusters  
‚úÖ Robusto para dados com densidades vari√°veis

## 3. One-Class SVM

![One-Class SVM](https://scikit-learn.org/stable/_images/sphx_glr_plot_oneclass_001.png)

### Como Funciona

O **One-Class SVM** √© uma adapta√ß√£o do SVM (Support Vector Machine) para detec√ß√£o de anomalias. Ele aprende uma **fronteira de decis√£o** que engloba a maioria dos dados normais.

O algoritmo funciona da seguinte forma:

1. **Mapeamento**: Projeta os dados em um espa√ßo de alta dimensionalidade usando uma fun√ß√£o **kernel** (RBF)
2. **Hiperplano**: Encontra um hiperplano que separa os dados da origem com **margem m√°xima**
3. **Fronteira**: Define uma "regi√£o normal" baseada na dist√¢ncia ao hiperplano
4. **Classifica√ß√£o**: Pontos fora dessa regi√£o s√£o classificados como anomalias

### Intui√ß√£o

Imagine uma "bolha" flex√≠vel que envolve os dados normais. O One-Class SVM ajusta essa bolha para:
- **Incluir** a maioria dos pontos (dados normais)
- **Excluir** pontos isolados ou distantes (anomalias)

### Par√¢metros Principais

| Par√¢metro | Valor Usado | Descri√ß√£o |
|-----------|-------------|-----------|
| `nu` | 0.015 | Limite superior da fra√ß√£o de anomalias (~1.5%) |
| `kernel` | 'rbf' | Fun√ß√£o kernel (Radial Basis Function) |
| `gamma` | 'scale' | Coeficiente do kernel (autom√°tico) |

### Vantagens
‚úÖ Fronteira de decis√£o flex√≠vel (n√£o-linear)  
‚úÖ Funciona bem quando h√° poucos exemplos de anomalias  
‚úÖ Baseado em teoria matem√°tica s√≥lida (otimiza√ß√£o convexa)

## 4. Isolation Forest (Deep) - Variante

![Isolation Forest Deep](https://scikit-learn.org/stable/_images/sphx_glr_plot_isolation_forest_002.png)

### Como Funciona

O **Isolation Forest (Deep)** √© uma variante do Isolation Forest padr√£o com o par√¢metro `max_samples` ajustado para criar √°rvores mais diversificadas.

### Diferen√ßa para o Isolation Forest Padr√£o

| Aspecto | Isolation Forest | Isolation Forest (Deep) |
|---------|------------------|-------------------------|
| `max_samples` | 'auto' (256 ou n) | 0.8 (80% dos dados) |
| Amostragem | Menor, mais r√°pido | Maior, mais diverso |
| √Årvores | Mais similares | Mais variadas |
| Generaliza√ß√£o | Boa | Potencialmente melhor |

### Por que usar esta variante?

Ao usar **80% dos dados** para construir cada √°rvore (ao inv√©s de uma amostra menor):

1. **Maior diversidade**: Cada √°rvore "v√™" mais dados
2. **Melhor cobertura**: Menos chance de perder padr√µes raros
3. **Trade-off**: Ligeiramente mais lento, mas pode capturar anomalias diferentes

### Par√¢metros Principais

| Par√¢metro | Valor Usado | Descri√ß√£o |
|-----------|-------------|-----------|
| `contamination` | 0.015 | Propor√ß√£o esperada de anomalias (~1.5%) |
| `max_samples` | 0.8 | Usa 80% dos dados em cada √°rvore |
| `random_state` | 42 | Semente para reprodutibilidade |

### Vantagens
‚úÖ Captura padr√µes que o IF padr√£o pode perder  
‚úÖ Ensemble mais robusto  
‚úÖ √ötil como modelo complementar para compara√ß√£o

# Fun√ß√µes de Indicadores T√©cnicos

Nesta se√ß√£o, definimos fun√ß√µes para calcular indicadores t√©cnicos amplamente utilizados em an√°lise financeira:

- **RSI (Relative Strength Index)**: Mede a velocidade e magnitude das mudan√ßas de pre√ßo. Valores acima de 70 indicam **sobrecompra** e abaixo de 30 indicam **sobrevenda**.

- **MACD (Moving Average Convergence Divergence)**: Indicador de momentum que mostra a rela√ß√£o entre duas m√©dias m√≥veis exponenciais. √ötil para identificar revers√µes de tend√™ncia.

- **Bollinger Bands (Banda Superior)**: Faixa superior calculada como m√©dia m√≥vel + 2 desvios padr√£o. Pre√ßos acima dessa banda podem indicar rompimentos ou anomalias.

A fun√ß√£o `advanced_features()` aplica engenharia de features criando:
- Indicadores t√©cnicos (RSI, MACD, Bollinger)
- M√©tricas de volume e volatilidade
- Features bin√°rias de anomalia (RSI extremo, picos de volume, gaps de pre√ßo)
- Features temporais (hora do dia, dia da semana)

```python
def calculate_rsi(series, period=14):
    delta = series.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=period).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=period).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

def calculate_macd(series, fast=12, slow=26):
    ema_fast = series.ewm(span=fast).mean()
    ema_slow = series.ewm(span=slow).mean()
    return ema_fast - ema_slow

def calculate_bollinger_upper(series, window=20, num_std=2):
    rolling_mean = series.rolling(window).mean()
    rolling_std = series.rolling(window).std()
    return rolling_mean + (rolling_std * num_std)

def advanced_features(df):
    df = df.copy()
    
    df['RSI'] = calculate_rsi(df['Close'])
    df['MACD'] = calculate_macd(df['Close'])
    df['BB_upper'] = calculate_bollinger_upper(df['Close'])
    
    df['Volume_Change'] = df['Volume'].pct_change()
    df['Volatility'] = df['Close'].rolling(24).std()
    df['Price_Change'] = df['Close'].pct_change()
    
    df['RSI_Extreme'] = (df['RSI'] > 80) | (df['RSI'] < 20)
    df['Volume_Spike'] = df['Volume_Change'] > df['Volume_Change'].quantile(0.95)
    df['Price_Gap'] = abs(df['Price_Change']) > df['Price_Change'].quantile(0.95)
    
    df['Hour'] = df.index.hour
    df['DayOfWeek'] = df.index.dayofweek

    return df
```

## Pr√©-processamento e Cria√ß√£o de Labels

Aplicamos o mesmo pipeline de pr√©-processamento:
1. Remo√ß√£o de valores infinitos
2. Preenchimento de valores ausentes
3. Clipping de extremos (percentis 1% e 99%)
4. Cria√ß√£o de labels sint√©ticos baseados em crit√©rios estat√≠sticos
5. Normaliza√ß√£o com RobustScaler

```python
features = ['Close', 'Volume', 'RSI', 'MACD', 'BB_upper', 'Volume_Change',
           'Volatility', 'Hour', 'DayOfWeek']
X_btc = btc[features].copy()

print("=== TRATAMENTO DE DADOS ===")
print("1. Removendo infinitos...")
X_btc.replace([np.inf, -np.inf], np.nan, inplace=True)

print("2. Preenchendo NaN...")
X_btc.fillna(0, inplace=True)

print("3. Clipping extremos...")
for col in X_btc.columns:
    lower = X_btc[col].quantile(0.01)
    upper = X_btc[col].quantile(0.99)
    X_btc[col] = X_btc[col].clip(lower, upper)

btc['Anomaly_Label'] = (
    (abs(btc['Price_Change']) > abs(btc['Price_Change']).quantile(0.99)) |
    (btc['RSI'] > 85) | (btc['RSI'] < 15) |
    (btc['Volume_Change'] > btc['Volume_Change'].quantile(0.98))
).astype(int)

scaler = RobustScaler()
X_btc_scaled = scaler.fit_transform(X_btc)

print(f"\nFeatures finais: {X_btc_scaled.shape}")
print(f"Anomalias reais: {btc['Anomaly_Label'].sum()} ({btc['Anomaly_Label'].mean():.1%})")
print("‚úÖ Pr√©-processamento conclu√≠do!")
```

## Treinamento dos Modelos

Utilizamos os mesmos 4 algoritmos de detec√ß√£o de anomalias:
- **Isolation Forest**: Isola anomalias usando √°rvores de decis√£o
- **Local Outlier Factor (LOF)**: Compara densidade local entre vizinhos

# Detec√ß√£o de Anomalias no TrumpCoin (TRUMP-USD)

Nesta se√ß√£o, aplicamos os mesmos algoritmos de detec√ß√£o de anomalias utilizados no Bitcoin para analisar o **TrumpCoin (TRUMP-USD)**, uma memecoin com alta volatilidade e comportamento de mercado diferente das criptomoedas tradicionais.

## Download e Prepara√ß√£o dos Dados

Baixamos os dados hist√≥ricos do TrumpCoin usando a mesma metodologia aplicada ao Bitcoin:
- **Ticker**: TRUMP-USD
- **Per√≠odo**: M√°ximo dispon√≠vel
- **Intervalo**: 1 hora

```python
print("=== DETEC√á√ÉO DE ANOMALIAS TRUMPCOIN ===\n")

# Download dos dados do TrumpCoin
trump_raw = yf.download('TRUMP-USD', period='max', interval='1h')

# Aplicar engenharia de features
trump = advanced_features(trump_raw)

print(f"\nDataset: {trump.shape}")
print(f"Per√≠odo: {trump.index.min():%Y-%m-%d} at√© {trump.index.max():%Y-%m-%d}")
print(f"Total de registros: {len(trump)}")
print("\nFeatures criadas:")
print([col for col in trump.columns if col not in trump_raw.columns])
```

## Pr√©-processamento e Cria√ß√£o de Labels

Aplicamos o mesmo pipeline de pr√©-processamento:
1. Remo√ß√£o de valores infinitos
2. Preenchimento de valores ausentes
3. Clipping de extremos (percentis 1% e 99%)
4. Cria√ß√£o de labels sint√©ticos baseados em crit√©rios estat√≠sticos
5. Normaliza√ß√£o com RobustScaler

```python
# Sele√ß√£o de features
features_trump = ['Close', 'Volume', 'RSI', 'MACD', 'BB_upper', 'Volume_Change',
                  'Volatility', 'Hour', 'DayOfWeek']
X_trump = trump[features_trump].copy()

print("=== TRATAMENTO DE DADOS - TRUMPCOIN ===")
print("1. Removendo infinitos...")
X_trump.replace([np.inf, -np.inf], np.nan, inplace=True)

print("2. Preenchendo NaN...")
X_trump.fillna(0, inplace=True)

print("3. Clipping extremos...")
for col in X_trump.columns:
    lower = X_trump[col].quantile(0.01)
    upper = X_trump[col].quantile(0.99)
    X_trump[col] = X_trump[col].clip(lower, upper)

# Cria√ß√£o de labels sint√©ticos
trump['Anomaly_Label'] = (
    (abs(trump['Price_Change']) > abs(trump['Price_Change']).quantile(0.99)) |
    (trump['RSI'] > 85) | (trump['RSI'] < 15) |
    (trump['Volume_Change'] > trump['Volume_Change'].quantile(0.98))
).astype(int)

# Normaliza√ß√£o
scaler_trump = RobustScaler()
X_trump_scaled = scaler_trump.fit_transform(X_trump)

print(f"\nFeatures finais: {X_trump_scaled.shape}")
print(f"Anomalias sint√©ticas: {trump['Anomaly_Label'].sum()} ({trump['Anomaly_Label'].mean():.1%})")
print("‚úÖ Pr√©-processamento conclu√≠do!")
```

- **Isolation Forest**: Isola anomalias usando √°rvores de decis√£o
- **Local Outlier Factor (LOF)**: Compara densidade local entre vizinhos
- **One-Class SVM**: Aprende fronteira ao redor dos dados normais
- **Isolation Forest (Deep)**: Variante com maior diversidade nas √°rvores

```python
print("=== TREINAMENTO DOS 4 MODELOS - TRUMPCOIN ===")

models_trump = {
    'Isolation Forest': IsolationForest(contamination=0.015, random_state=42, n_jobs=-1),
    'Local Outlier Factor': LocalOutlierFactor(n_neighbors=25, contamination=0.015),
    'One-Class SVM': OneClassSVM(nu=0.015, kernel='rbf', gamma='scale'),
    'Isolation Forest (Deep)': IsolationForest(contamination=0.015, max_samples=0.8, random_state=42)
}

predictions_trump = {}
for name, model in models_trump.items():
    print(f"Treinando {name}...", end=" ")
    preds = model.fit_predict(X_trump_scaled)
    anomalies = np.where(preds == -1)[0]
    predictions_trump[name] = anomalies
    print(f"{len(anomalies)} anomalias")

print("\n‚úÖ Todos os modelos treinados para TrumpCoin!")
```
